```julia
using JLD2
using MPSTime
```

```julia
files_raw = readdir("../ImputationLoss()_BBO_adaptive_de_rand_1_bin(:adaptive_de_rand_1_bin)_f=30_cv=5_iters=50_tmp")
```

```julia
function get_fold(fname::String)
    i_start = 2
    i_end = i_start
    while fname[i_end] !== '.'
        i_end +=1
    end
    return parse(Int, fname[i_start:i_end-1]) 

end
```

```julia
res = Dict{Int, Any}()
for file in files_raw
    if length(file) > 5 && file[end-4:end] == ".jld2"
        fold = get_fold(file)
        @load file res_iter
        res[fold] = res_iter
    end

end
```
