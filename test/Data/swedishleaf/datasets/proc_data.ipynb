```julia
using Pkg
Pkg.activate("../../../Project.toml")

using JLD2
using DataFrames
using DelimitedFiles
using Plots
```

```julia
# read data read
test_path = "SwedishLeaf_TEST.txt"
train_path = "SwedishLeaf_TRAIN.txt"
test_series = 625
train_series = 500
ts_length = 128
```

```julia
X_test_labelled = readdlm(test_path)
X_train_labelled = readdlm(train_path)


# fixed array size so itll break if the data isnt layed out as expected
X_test = Matrix{Float64}(undef, test_series, ts_length)
X_train = Matrix{Float64}(undef, train_series, ts_length);

y_test = Vector{Int}(undef, test_series)
y_train = Vector{Int}(undef, train_series)


X_test .= X_test_labelled[:, 2:end] 
y_test .= X_test_labelled[:,1];

X_train .= X_train_labelled[:, 2:end] 
y_train .= X_train_labelled[:,1];
```

```julia
plot(X_train[5,:])
```

```julia
jldopen("SwedishLeaf.jld2", "w") do f
    f["X_train"] = X_train
    f["y_train"] = y_train
    f["X_test"] = X_test
    f["y_test"] = y_test
end
```
