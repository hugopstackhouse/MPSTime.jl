```julia
using JLD2
using CSV
using DataFrames
using Plots
```

```julia
label_mapping = Dict(
    0 => :ContactEB,
    1 => :DetachedEB,
    2 => :DeltaScVar,
    3 => :GammaDorVar,
    4 => :NonVar,
    5 => :RotVar,
    6 => :RRLyrVar
);
```

```julia
df = CSV.read("KeplerLightCurves.csv", DataFrame; delim = ',', header=false);
```

```julia
dataset = Matrix(df);
dataset_final = dataset[:, 1:end-1]
labels = Int.(dataset[:, end]) .- 1;
labels_mapped = [label_mapping[l] for l in labels];
```

# Inspect Class Distribution

```julia
histogram(labels, label = "", xlabel = "Class", ylabel = "Count")
```

# Inspect some examples

```julia
pal = palette(:tab10);
```

# Class 0 - Contact Eclipsing Binary

```julia
all_contact_eb = findall(x -> x .== :ContactEB, labels_mapped);
plot(dataset_final[all_contact_eb[100], :], label="", title="Contact Eclipsing Binary", xlabel="Time (samples)", ylabel="X", c=pal[2])
#savefig("exploration/contact_eclip_binary100.svg")
```
