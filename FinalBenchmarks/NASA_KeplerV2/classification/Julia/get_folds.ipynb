```julia
folds = Dict()
for f in 1:30
    train_idxs = split_idxs[f][1]
    train_idxs .-= 1
    test_idxs = split_idxs[f][2]
    test_idxs .-=1
    train_test = Dict("train" => train_idxs, "test" => test_idxs)
    folds[f-1] = train_test
end
```

```julia
open("KeplerBenchFoldsPython.json", "w") do f
    JSON.print(f, folds)
end
```
