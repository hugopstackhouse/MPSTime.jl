```julia
using JSON
using Plots
using JLD2
```

```julia
mps_f = jldopen("epilepsy2_30fold_mps.jld2", "r");
mps_results = read(mps_f, "fold_accs");
ict_f = JSON.parsefile("epilepsy2_inceptiontime_30fold.json")
ict_results = ict_f["per_fold_accs"];
nn_results = JSON.parsefile("epilepsy2_nndtw_30fold.json");
println("$(mean(mps_results)) | $(std(mps_results))")
println("$(mean(ict_results)) | $(std(ict_results))")
println("$(mean(nn_results)) | $(std(nn_results))")
mean_mps = mean(mps_results)
mps_std = 1.96 * std(mps_results)/sqrt(30)
mean_ict = mean(ict_results)
ict_std = 1.96 * std(ict_results)/sqrt(30)
mean_nn = mean(nn_results)
nn_std = 1.96 * std(nn_results)/sqrt(30)
```

```julia
bar([mean_mps, mean_nn, mean_ict], c=:green,
    yerr=[mps_std nn_std ict_std], ylims=(0.9, 0.99))
```
